```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import re
```

```python
target_url = "https://www.magicbricks.com/property-for-sale/commercial-real-estate?&cityName=Coimbatore&proptype=Commercial-Shop,Commercial-Showroom"
web_page = requests.get(target_url)
```

```python
html_source_code = BeautifulSoup(web_page.content, 'html.parser')
target_list = html_source_code.find_all('div', class_="mb-srp__list")
print(target_list)
```

```python
with open ('Commercial Properties for Sale in Coimbatore.csv','w',encoding='utf8',newline='') as f:
  thewriter = writer(f)
  header = ['Property_Title', 'Furnishing_Status', 'Area',  'Price']
  thewriter.writerow(header)
  for list in target_list:
    title = list.find('h2', class_='mb-srp__card--title').text.replace('\n','')
    x_list = list.find_all('div', class_='mb-srp__card__summary--value')
    for i in reversed(x_list):
        if re.search('Furnished', i.text):
            furnishing_status = 'Furnished'
            break
        elif re.search('Unfurnished', i.text):
            furnishing_status = 'Unfurnished'
            break
        else:
            furnishing_status = 'Contact Owner'

    area = list.find('div', class_='mb-srp__card__summary--value').text.replace('\n','')
    price = list.find('div', class_='mb-srp__card__price--amount').text
    
    info = [title,furnishing_status,area,price]
    thewriter.writerow(info)
```
